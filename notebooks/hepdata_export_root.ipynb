```python
import copy
from yaml import load, dump

try:
    from yaml import CLoader as Loader, CDumper as Dumper
except ImportError:
    from yaml import Loader, Dumper

template_1D = {
    'independent_variables': [
        {
            'header': {"name": "x-axis", "units": "GeV"},
            'values': [],
        },
        ],
    'dependent_variables': [
        {
            'header': {"name": "actual values", "units": "GeV"},
            'values': [],
        }

    ]
}
    
def contour_to_hepdata(contour, name='test', xaxis="m_{NLSP}", yaxis="m_{LSP}"):
    tmp = copy.deepcopy(template_1D)
    tmp['independent_variables'][0]['header']['name'] = xaxis
    tmp['dependent_variables'][0]['header']['name'] = yaxis
    X, Y = contour.values()
    for x, y in zip(X,Y):
        tmp['independent_variables'][0]['values'].append({'value': "%.4g"%x})
        tmp['dependent_variables'][0]['values'].append({'value': "%.4g"%y})

    f_out = name+'.yaml'
    with open(f_out, 'w') as f:
        dump(tmp, f, Dumper=Dumper)

    return f_out
```

```python
for name in f_in.keys():
    if name.count('TChiWHObservedExcludedXsec'): continue
    contour_to_hepdata(f_in[name], name=name.replace(';1',''))
```

```python
template_2D = {
    'independent_variables': [
        {
            'header': {"name": "x-axis", "units": "GeV"},
            'values': [],
        },
        {
            'header': {"name": "y-axis", "units": "GeV"},
            'values': [],
        },],
    'dependent_variables': [
        {
            'header': {"name": "actual values", "units": "pb"},
            'values': [],
        }

    ]
}
```

```python
h_exp_2D.axes[0].centers()
```

```python
obs_yaml = copy.deepcopy(template_2D)

for ix, x in enumerate(h_exp_2D.axes[0].centers()):
    for iy, y in enumerate(h_exp_2D.axes[1].centers()):
        limit = h_exp_2D.counts()[ix, iy]
        if limit > 0:
            #print (x, y, limit)
            # observed
            obs_yaml['independent_variables'][0]['values'].append({'value': str(x)})
            obs_yaml['independent_variables'][1]['values'].append({'value': str(y)})
            obs_yaml['dependent_variables'][0]['values'].append({'value': "%.3g"%(limit)})
            
obs_yaml['independent_variables'][0]['header']['name'] = "m_{NLSP}"
obs_yaml['independent_variables'][1]['header']['name'] = "m_{LSP}"
obs_yaml['dependent_variables'][0]['header']['name'] = "95% CL upper limit on cross section"

f_out = 'observed_limit.yaml'
with open(f_out, 'w') as f:
    dump(obs_yaml, f, Dumper=Dumper)
```
